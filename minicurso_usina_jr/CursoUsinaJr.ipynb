### Criando classes

Outro conceito importante em Python é o de classe. Classes são objetos que contêm variáveis e métodos (outro nome pra funções, mas no contexto de classes) próprios num "objeto" auto-contido com propriedades semelhantes uns aos outros, mas que podem ter valores diferentes. Vamos ver um exemplo.

```python
class UNI(): #Objeto "Universidade"
    '''
    Mais um exemplo de documentação (dessa vez sem muito uso, não façam isso!)
    '''
    
    def __init__(self, nome, melhor_curso = 'engenharia metalúrgica'): #Este nome __init__ é especial! 
        '''
        Inicialização do objeto, atribuindo o nome e o melhor curso da universidade.
        '''
        
        self.nome = nome #Toda "Universidade" tem nome, mas cada uma tem um diferente.
        self.melhor_curso = melhor_curso #O mesmo aqui
        self.matriculas = []
        
    def imprimir_melhor_curso(self):
        '''
        Podemos documentar aqui também. Nessa função imprimimos o
        nome do melhor curso da universidade que definimos antes.
        '''
        
        print(f'O melhor curso é o curso de {self.melhor_curso}!')
    def fazer_matricula_e_checar_matriculas(self,numero_matricula):
        '''
        Adiciona 'numero_matricula' à lista de matrículas. Numa aplicação real obviamente
        deveríamos ter outros cuidados aqui (e.g., não ter repetições, checar se o argumento
        da função realmente é um número, etc.)
        '''
        
        self.matriculas.append(numero_matricula)
        print(f"A matrícula {numero_matricula} foi registrada!")
        return self.matriculas.copy()
```

```python
Universidade_1 = UNI(nome='UFMG') #Instanciando um objeto da classe 'UFMG'

Universidade_1.imprimir_melhor_curso() #Usando um método dessa classe
```

```python
#Além de chamar os métodos, podemos acessar os valores das variáveis do objeto "Universidade"

print(Universidade_1.nome)
```

```python
Universidade_2 = UNI(nome='USP',melhor_curso='ciências moleculares')

Universidade_2.imprimir_melhor_curso()
```

```python
matricula_0 = Universidade_1.fazer_matricula_e_checar_matriculas('00001')
print(matricula_0)
matricula_1 = Universidade_1.fazer_matricula_e_checar_matriculas('00002')
print(matricula_1)
```

### O jeito "Pythonico" de programar

A estrutura flexível da linguagem permite fazer algumas coisas que em outras linguagens poderiam ser mais complicadas. Usar essa estrutura da maneira certa é o chamado jeito "Pythonico" de programar. Pegar esses truques é questão de prática, estudo e tempo.

```python
#Por exemplo, remover elementos repetidos de uma lista

x = [1,2,3,4,10,2,7,8,2,2]

#Jeito 'intuitivo' (não necessariamente o mais rápido ou melhor)

unique = []
for i in range(len(x)): #len(x) retorna o número de elementos da lista
    if x[i] not in unique:
        unique.append(x[i])

print(unique)

#Jeito 'pythonico'

unique = list(set(x)) #Transforma num conjunto, depois volta pra uma lista
print(unique)
```

- Exemplo (2) - Funções bidimensionais

Visualizando a função $z = f(x,y) = \sin{(e^{x}+\ln{|y|})}$.

```python
x = np.linspace(-3,3,200) #Cria um vetor unidimensional
y = np.linspace(-3,3,200) #Cria outro vetor unidimensional

x, y = np.meshgrid(x, y) #Junta ambos em um grid de pontos bidimensional

z = np.sin(np.exp(x)+np.log(np.abs(y)))

plt.contourf(x,y,z,cmap='Blues') #Contour 'filled', por isso o 'f'
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.xlabel('x',fontsize=15)
plt.ylabel('y',fontsize=15)
plt.title('z = f(x,y)',fontsize=15)
plt.colorbar()
plt.show()
```

- Exemplo (3) - Calculando o valor de $\pi$. 

Podemos calcular o valor de $\pi$ distribuindo uniformemente valores em um quadrado de lado 2 e contando o que cai dentro de um círculo de raio 1 no centro desse quadrado em relação ao total.

Como a área do quadrado será 4, a do círculo seria $\pi$, a razão $A_O/A_{\square} = \pi/4$.

(Isso que veremos é um exemplo simples de simulações de Monte Carlo).

```python
np.random.seed(0)

points = np.random.uniform(-1,1,size=(100000,2))
points[10]
```

```python
inside = 0
total = 0

color=[]

for point in points: #Iterando sobre os pontos da lista que definimos acima
    if (point[0]**2+point[1]**2) <= 1:
        inside+=1
        color.append('blue')
    else:
        color.append('red')
    total+=1
    
    
print( 4 * inside/total)
```

```python
ax, fig = plt.subplots(1,1,figsize=(5,5))
plt.scatter(x=points[:,0],y=points[:,1],c=color)
plt.show()
```

## O scikit-learn, sua estrutura e problemas de exemplo

![](https://upload.wikimedia.org/wikipedia/commons/0/05/Scikit_learn_logo_small.svg)

- O scikit-learn é a biblioteca base pra quase toda modelagem em ciência de dados usando Python. Ele tem diversos modelos implementados de maneira eficiente e várias funções úteis pra tratar dados, além de permitir composições e regulagem dessas funções de maneira bastante eficiente.


- A estrutura geral do scikit-learn consiste em criar um 'estimador' (uma classe), que vai aprender algo sobre os dados fitando eles (com o método <em>.fit</em>) e usar essa informação seja pra predizer algo sobre novos dados (<em>.predict</em>) ou pra transformar dados (<em>.transform</em>).

### Uma aplicação básica: regressão linear


Achar uma função $\hat{y} = \hat{y}(x)$ linear que minimize os erros quadrados, ou seja, que minimize $L(\hat{y}) = \frac{1}{2N} \sum_{i}^N  (y_i - \hat{y}(x_i))^2$, onde $N$ é o número de pontos $(x_i, y_i)$ que temos nos dados.

![](https://upload.wikimedia.org/wikipedia/commons/b/b0/Linear_least_squares_example2.svg)

<sub>(Krishnavedala, tirado de https://en.wikipedia.org/wiki/Linear_least_squares)</sub>

```python
#Dividindo em dados de treino e teste

from sklearn.linear_model import LinearRegression #O modelo que vamos usar
from sklearn.model_selection import train_test_split #Função pra dividir entre treino e teste os dados

lin_reg = LinearRegression() #Instancia um estimador, um exemplo das classes que discutimos no começo

X, y = df['petal_length'].values, df['petal_width'].values

X, y = X.reshape(-1,1), y.reshape(-1,1) #Método dos arrays do np pra deixar os dados do jeito 'certo' pro sklearn

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.15,random_state=0) 

lin_reg.fit(X_train,y_train) #Fitando o modelo

y_pred = lin_reg.predict(X_test).reshape(-1)

plt.scatter(X_train,y_train,color='blue',label='Treino')
plt.plot(X_test,y_pred,color='red',label='Predito')
plt.scatter(X_test,y_test,color='yellow',label='Teste')
plt.title('Exemplo de regressão linear.')
plt.legend()
plt.show()
```

### Outra aplicação básica: classificando as espécies de flor

Vamos ver um modelo de classificação bastante usado tanto por si como sendo base de outros modelos: uma árvore de decisão. 

Uma árvore de decisão é um conjunto de regras "se, então" escolhido de maneira "inteligente" (i.e., pra classificar da melhor maneira possível os dados de treino que tivermos). Vamos mostrar isso graficamente
em breve.

Também vamos falar um pouco sobre overfitting e underfitting (variando a profundidade da árvore).

```python
#Classificação com uma árvore de decisão

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_score, accuracy_score, recall_score, f1_score

features=df.columns[:-1] #As características que vamos usar
target=df.columns[-1] #O que queremos prever

X, y = df[features].values, df[target].values

dt = DecisionTreeClassifier(random_state=15) #Instanciando a árvore de decisão

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)

dt.fit(X_train,y_train)

y_pred = dt.predict(X_test)

print(f"Precisão: {precision_score(y_test,y_pred,average='weighted')}")
print(f"Recall: {recall_score(y_test,y_pred,average='weighted')}")
print(f"Acurácia: {accuracy_score(y_test,y_pred)}")
print(f"F1 score: {f1_score(y_test,y_pred,average='weighted')}")
```

### O que é uma rede neural?

![](https://upload.wikimedia.org/wikipedia/commons/4/46/Colored_neural_network.svg)

<sub>(Glosser.ca, tirado de https://en.wikipedia.org/wiki/Artificial_neural_network)</sub>

Antes disso precisamos pré-processar os dados pra deixar eles no formato certo.

```python
X_train = np.array([X_train_orig[i].reshape(-1) for i in range(len(X_train_orig))])
y_train = to_categorical(y_train_orig)

X_test = np.array([X_test_orig[i].reshape(-1) for i in range(len(X_test_orig))])
y_test = to_categorical(y_test_orig)
```

```python
#Vejamos os dados como ficaram:

X_train[0] #Um vetor com 784 entradas
```

```python
y_train[0]
```

Em geral precisamos sempre fazer algum tipo de tratamento nos dados antes de passar eles pra rede. Um tratamento bem comum é trazer todos os números pra números da ordem de ~ 1. Isso ajuda o modelo a aprender melhor por conta das operações matemáticas que ele tem que realizar pra aprender.

Aqui a conversão que nós fazemos é:

- $x_i \to \frac{x_i-\bar{x}}{\sigma}$

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() #Novamente o mesmo padrão: instanciar, fitar e transformar
X_train = scaler.fit_transform(X_train) #Fita e transforma os dados de treino
X_test = scaler.transform(X_test) #Somente transforma os dados de teste
```

```python
X_train[0]
```

### Criando a rede

```python
#keras.backend.clear_session() #Limpa a memória antes de rodar de novo

x_in = Input(shape=784)
x = Dense(16,activation='linear')(x_in)
y_out = Dense(10,activation='softmax')(x)

model = Model(inputs=x_in,outputs=y_out)
model.compile(optimizer='sgd',loss='categorical_crossentropy')
model.summary()
```

### Treinando a rede

```python
history=model.fit(X_train,y_train,epochs=20,batch_size=128,validation_data=(X_test,y_test))
```

### Avaliando os resultados

```python
plt.plot(history.history['loss'],label='loss')
plt.plot(history.history['val_loss'],label='val_loss')
plt.legend()
plt.show()
```

```python
y_pred=model.predict(X_test).argmax(axis=1) #Prevendo os dados de teste
y_pred[:10]
```

### Usando uma arquitetura mais apropriada pro problema

```python
X_train = X_train_orig.reshape(-1,28,28,1) #Transformamos os dados em um vetor com N dados de entrada, 28 x 28 x 1
y_train = to_categorical(y_train_orig)

X_test = X_test_orig.reshape(-1,28,28,1)
y_test = to_categorical(y_test_orig)

x_in = Input(shape=(28,28,1))
x = Conv2D(kernel_size=2,filters=8,activation='relu',padding='same')(x_in) #Usando outro tipo de "camada" pra rede
x = MaxPool2D(2)(x)
x =  Conv2D(kernel_size=2,filters=16,activation='relu',padding='same')(x)
x = MaxPool2D(2)(x)
x = Flatten()(x)
x =  Dense(16,activation='relu')(x)
y_out = Dense(10,activation='softmax')(x)

model = Model(inputs=x_in,outputs=y_out)
model.compile(optimizer='adam',loss='categorical_crossentropy')
model.summary()
```

```python
model.fit(X_train,y_train,epochs=15,batch_size=128, validation_data=(X_test,y_test))
```

```python
accuracy_score(y_test_orig,model.predict(X_test).argmax(axis=1))
```
